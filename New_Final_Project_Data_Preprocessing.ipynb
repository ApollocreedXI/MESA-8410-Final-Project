Notice how we have some catergorical variables that can make it difficult to an analysis on. What we can do is encode the variables to give numerical values to the ordinal variables.

##Will begin encoding variables for use in further analysis.

I want to note that I did use external sources and gen AI to introduce to me the library that can encode variables and an example of how it is used. At this point in the class we had not gone over encoding. This is the main reason I utilized this method.

```python
from sklearn.preprocessing import OrdinalEncoder

# Define the desired order for the variables

desired_wcase = ["too little attention", "enough attention", "too much attention"]

desired_effective = ["very ineffective", "ineffective", "neither effective nor ineffective" , "effective", "very effective"]

#Define two different encoders

encoder_wcase = OrdinalEncoder(categories=[desired_wcase])

encoder_effective = OrdinalEncoder(categories=[desired_effective])

#encode the values for each variable

#creating lists for the loop that contains column names
wcase_vars = ["e_wcase", "b_wcase"]

effective_vars = ["e_effective", "b_effective"]

#loop for wcase columns
for col in wcase_vars:
  police_survey_merge[col + '_encoded'] = encoder_wcase.fit_transform(police_survey_merge[[col]])

#loop for effective columns

for col in effective_vars:
  police_survey_merge[col + '_encoded'] = encoder_effective.fit_transform(police_survey_merge[[col]])
```

```python
#Lets check to see if we get the values we expect
enc = "_encoded"
check_df = police_survey_merge[["e_wcase","e_wcase_encoded","b_wcase", "b_wcase_encoded","e_effective", "e_effective_encoded", "b_effective", "b_effective_encoded"]]

check_df.head(15)

#Providing ordinal values for referance

#desired_wcase = ["too little attention", "enough attention", "too much attention"]

#desired_effective = ["very ineffective", "ineffective", "neither effective nor ineffective" , "effective", "very effective"]
```

I will now check for summary statistics to see if there are any extraneuos values

```python
#Will check for summmary statistics

citizen_survey_df.describe()
```

There are a couple of things to noticed from here


*   About 50% of the data had someone else answer that was not the original person surveyed or did not have someone to respond at endline (attrited Variable)
* One could argue that we should only take survey responses from those who answered at baseline and endline. This would be the only way to make any casual claims.



```python
# We will filter the data set for those women respondents that answered both at baseline and endline
# In particular for those questiopns pertaining to perceived effectivness and safety

citizen_survey_df = citizen_survey_df[(citizen_survey_df.b_safety.notna()) & (citizen_survey_df.e_safety.notna())]

citizen_survey_df = citizen_survey_df[(citizen_survey_df.b_pol_handling.notna()) & (citizen_survey_df.e_pol_handling.notna())]

citizen_survey_df = citizen_survey_df[(citizen_survey_df["member_gender"] == "female") & (citizen_survey_df["e_gender"] == "Female")]
```

```python
citizen_survey_df.shape
```

```python
#Check percent women who responded at baseline and did not responded at endline

percentage = citizen_survey_df.value_counts('attrited', normalize= True)

percentage = percentage.reset_index()

display(percentage)
```

It seems like the filtering worked as we should not have any attrited values. We removed all endline values that had blank endline responses and any respondents that had a different gender response at endline.

```python
citizen_survey_df.head(15)
```

```python
citizen_survey_df.info()
```

```python
#Encode the remaining categorical variables

from sklearn.preprocessing import OrdinalEncoder


#Define encoder

encoder = OrdinalEncoder()

#encode the values for each variable

citizen_survey_df["e_gender_encoded"] = encoder.fit_transform(citizen_survey_df[["e_gender"]])

citizen_survey_df["member_gender_encoded"] = encoder.fit_transform(citizen_survey_df[["member_gender"]])

citizen_survey_df["urban_encoded"] = encoder.fit_transform(citizen_survey_df[["urban"]])

citizen_survey_df["treatment_encoded"] = encoder.fit_transform(citizen_survey_df[["treatment"]])
```

```python
#Checking if encoding worked properly. Will create a dataframe to make side-by-side comparisons
check_df = citizen_survey_df[["e_gender", "e_gender_encoded", "member_gender", "member_gender_encoded", "urban","urban_encoded", "treatment", "treatment_encoded"]]

check_df.head(20)
```

It looks like the encoding worked correctly as each binary variable is assigned a numerical value for the responses.

```python
#Encode the categorical variables

from sklearn.preprocessing import OrdinalEncoder

#Set variable equal to OrdinalEncoder class

en = OrdinalEncoder()

#Encode variables

admin_long_data_df["urban_encoded"] = en.fit_transform(admin_long_data_df[["urban"]])

admin_long_data_df["treatment_encoded"] = en.fit_transform(admin_long_data_df[["treatment"]])
```

```python
admin_long_data_df.head()
```

```python
admin_long_data_df.describe(include="all")
```

The month variable is string variable. Lets see if we can extract the month and year in these rows

```python
#lets check out the formatting of the month variable

month_counts = admin_long_data_df.value_counts("month")

month_counts = month_counts.reset_index()

display(month_counts)
```

The variable month is in the format YYYY-MM-DD. We can extract the dates to create two new integer variables. Another thing to note, the data shows the first day of the month for the month ending activity. However, in investigating it seems like this is is just a  default notation when one uses month and year.

```python

#extracting year from the month column


admin_long_data_df["year_YYYY"] = [x[:4] for x in admin_long_data_df["month"]]

admin_long_data_df["month_MM"] = [x[5:7] for x in admin_long_data_df["month"]]

#converting strings into integers

admin_long_data_df["year_YYYY"] = [int(x) for x in admin_long_data_df["year_YYYY"]]

admin_long_data_df["month_MM"] = [int(x) for x in admin_long_data_df["month_MM"]]

admin_long_data_df.info()

```

```python
#making sure we did everything correct

admin_long_data_df[["month", "month_MM", "year_YYYY"]]
```

```python
#export dataframe

admin_long_data_df.to_csv("admin_long_cleaned.csv")
```
